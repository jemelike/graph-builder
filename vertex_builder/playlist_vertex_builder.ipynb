```python
from __future__  import print_function  # Python 2/3 compatibility

import pkg_resources
import json
import pandas as pd
import boto3
pkg_resources.get_distribution("boto3").version
```

```python
s3 = boto3.client('s3')

response = s3.get_object(
    Bucket='bearly-graph',
    Key='src/mpd.slice.0-999.json',
)
# Extracts body of the 
body = response['Body']
str_body = body.read()
s3_obj_content = json.loads(str_body)
playlists = s3_obj_content['playlists']
```

```python
playlist_list = []

for playlist in playlists:  
    
    playlist_dict = {}
    playlist_dict['~id'] = f"playlist_{playlist['pid']}"
    playlist_dict['name:String'] =  playlist['name']
    playlist_dict['modified_at:Int'] =  playlist['modified_at']
    playlist_dict['num_albums:Int'] =  playlist['num_albums']
    playlist_dict['num_tracks:Int'] =  playlist['num_tracks']
    playlist_dict['num_followers:Int'] =  playlist['num_followers']
    playlist_dict['num_edits:Int'] =  playlist['num_edits']
    playlist_dict['duration_ms:Int'] =  playlist['duration_ms']
    playlist_dict['num_artists:Int'] =  playlist['num_artists']
    playlist_dict['label'] =  'playlist'
    playlist_list.append(playlist_dict)
    

df = pd.DataFrame.from_dict(playlist_list)

    
```

```python
b_string = df.to_csv().encode('utf-8')
s3.put_object(Body=b_string, Bucket='bearly-graph', Key='vertex/version=2/playlist_vertex_0.csv')
```
