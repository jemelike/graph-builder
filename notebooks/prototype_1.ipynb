```python
# Setup connection to graph
graph = Graph()
remoteConn = DriverRemoteConnection('wss://neptunedbcluster-ojo9zmfsbp7d.cluster-cqlijw7gc9bd.us-east-1.neptune.amazonaws.com:8182/gremlin','g')
g = graph.traversal().withRemote(remoteConn)
```

```python
# retrieves edges related to missy_elliot
try:
    vertices = g.V().toList()
except RuntimeError as runtime_error:
    print('Failed to execute query.')
    print(runtime_error)
```

```python
list_size = 5
i = 0 
randomly_selected_songs = []
max_index = len(vertices) - 1
while i <= list_size:
    x = randint(0,max_index)
    if vertices[x] in randomly_selected_songs:
        x = randint(0,max_index)
    else:
        randomly_selected_songs.append(vertices[x])    
        i += 1
```

```python
song_relations = [ ]

for v in randomly_selected_songs:
    relations = g.V(v.id).outE().outV().valueMap().toList()
    
    song_relations.append(relations)

song_relations
```
